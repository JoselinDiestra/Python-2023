## CAPITULO 1: VARIABLES Y TIPOS DE DATOS BASICOS

**Variable:** es un lugar designado en la memoria donde se guardan datos y luego estos datos se puede "llamar" utilizando el nombre de la variable.

**¿Cómo nombrar variables?** Se usan nombres abreviados que describan los valores a guardar. En caso el nombre de variable tenga más de una palabra se usa el "_" o sino se usan mayúsuculas en las primera letras, por ejemplo: precio_dolar o precioDolar

**Tipos de datos básicos en python:** 
- *Integer:* hace referencia a un valor numérico entero
- *Float:* hace referencia a un valor numérico con decimales
- *String:* hace referencia a un valor de texto y se coloca entre comillas simples (') o comillas dobles (")
- *Booleano:* hace referencia a un valor binario (True o False) o (1 y 0)

### Variable entera

```python
# definición de variables
# Variable entera

edad = 20 
print(edad)
type(edad)  # int: integer
```

### Variable float

```python
# variable con decimales (float)

precio_dolar = 3.8
print(precio_dolar)
type(precio_dolar)
```

```python
precio_aceite = 12.5
print(precio_aceite)
type(precio_aceite)
```

### Operaciones básicas

```python
# Operaciones matemáticas
cant_alumnos = 13
edad = 15

suma_edades = cant_alumnos*edad
print(suma_edades)
```

```python
cant_dolares = 2000
precio_compra = 2.5
precio_venta = 3.8

ganancia = cant_dolares*(precio_venta-precio_compra)
print(round(ganancia,0))
```

```python
# operación de potencia
a= 3
b= 5

c = a**b
print(c)
```
